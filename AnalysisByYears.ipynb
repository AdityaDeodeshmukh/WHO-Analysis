<font size="5.5">Analysis of WHO Life Expectancy Dataset</font>

This is an analysis of data published by the Global Health Observatory (GHO) data repository under World Health Organization (WHO). The goal of this analysis is to find if correlation exists between the life expectancy of a country and various factors such as vaccinations, Alcohol consumption, Expendeture on Healthcare etc. 

```python
#importing all the needed modules
import pandas as pd
import os
```

```python
df=pd.read_csv("Life Expectancy Data.csv")
df.head()
```

```python
df.describe()
```

Here we see that the mean life expectancy from 2000 to 2015 is 69.224932. Now, we shall analyse the mean and median life expectancy year by year from 2000 to 2015. 

```python
df1=df.copy()
```

```python
df2=df1.groupby(by="Year").mean()
df2=df2["Life expectancy "]
df2
```

```python
df2=df1.groupby(by="Year").median()
df2=df2["Life expectancy "]
df2
```

As we can see life expectancy, on average, has risen in the period of 2000 and 2015. The mean life expectancy has risen from 66.75 Years to 71.61 Years, which is a rise of About 7%. The median life expectancy has increased from 71 to 73.9, which is an increase of 4%

Now we shall perform analysis of other factors like adult mortality and infant deaths

```python
df2=df1.groupby(by="Year").mean()
df2=df2[["Adult Mortality","infant deaths"]]
df2
```

As we can see the number of infant deaths(per 1000) as well as the Adult Mortality has decreased drastically

Now let us analyse the life expectancy for developed and developing countries seperately

```python
df2=df[df["Status"]=="Developing"]
df2=df2.groupby(by="Year").mean()
df2[["Life expectancy ","Adult Mortality","infant deaths"]]
```

As we see, for developing countried, there is a significant increase in the life expectancy and a drastic decrease in adult mortality and the number of infant deaths(per 1000). 

```python
df2=df[df["Status"]=="Developed"]
df2=df2.groupby(by="Year").mean()
df2[["Life expectancy ","Adult Mortality","infant deaths"]]
```

We see a similar result when it comes to developed countries
